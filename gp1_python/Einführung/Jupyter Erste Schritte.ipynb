# Jupyterhub für das Grundpraktikum 1

## Erste Schritte

Für die Datenauswertung im Grundpraktikum bieten wir eine Jupyter Umgebung in OLAT an. Um die Umgebung zu starten, einfach auf den Button auf `Jupyter Hub starten` klicken.



Nach dem Klick sollte ein neuer Tab im Browser öffnen. Sollte das nicht der Fall sein, bitte die Einstellungen des Browsers überprüfen und das Öffnen von Tabs zulassen.


## Installation der Tutorials und Vorlagen

Die Tutorials und Vorlagen müssen in die Jupyter Installation integriert werden. Dazu muss ein Git-Repository heruntergeladen werden. Dazu muss auf der linken Seite, neben dem Dateibrowser der `GIT` reiter ausgewählt werden und der Button `clone a repository` geklickt werden.



Die URI des Repository zum Grundpraktikum ist [https://git.uibk.ac.at/c704271/gp1_python.git](https://git.uibk.ac.at/c704271/gp1_python.git)



Nach dem Klonen sollten sie einen Ordner `gp1_python` vorfinden, in dem die Tutorials und Vorlagen sind.


Auf der linken Seite der Umgebung befindet sich nun ein Datei-Browser mit dem Ordner `gp1_python`. Dieser Ordner enthält Tutorials und Vorlagen für die Datenauswertung. Achtung: Die Dteien existieren nur auf dem Server der Universität und nich auf dem Rechner.

  



Im Ordner `Einführung` befindet sich die Datei `Einführung_in_Python.ipynb`. 

Hierbei handelt es sich um eine speziell für Physikpraktika angepasste Einführung in Python. Diese Datei soll geöffnet, verstanden und ausgeführt werden. Die Einführung beruht auf dem Buch "Mit Jupyter durchs Physikpraktikum", das unter folgenden Link bezogen werden kann: [https://link.springer.com/book/10.1007/978-3-658-37723-6](https://link.springer.com/book/10.1007/978-3-658-37723-6)

# Jupyter lokal installieren

Die in OLAT integrierte Jupyter Umgebung hat zwei Nachteile:

- Man benötigt eine Internetverbindung um daran zu arbeiten
- Die Dateien werden nicht lokal auf dem jeweiligen Rechner gespeichert

Es gibt auch die Möglichkeit Jupyter lokal auf einem eigenen Rechner zu installieren. Dieser Schritt ist optional und nur nötig falls eine eigene Jupyter Umgebung gewünscht wird.

Dazu kann das Softwarepaket `Anaconda` verwendet werden: [https://www.anaconda.com/download](https://www.anaconda.com/download) Nach dem Download und der Installation kann Jupyter aus dem Anaconda Launcher gestartet werden.
