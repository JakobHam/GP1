Hier soll ein linearer Fit durchgeführt werden

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
```

```python
# Setze hier die Daten, die angepasst werden sollen ein:
array_x = np.array([1.1,2.2,3.1,4.3])
array_y = np.array([2.1,4.3,6.1,8.9])

# Hier die Unsicherheit (Standardabweichung)
array_y_std = np.array([1,1,1,1])
y_weights = 1/array_y_std
```

```python
params, cov_mat = np.polyfit(array_x,array_y,deg=1,w=y_weights , cov=True)
```

```python
a = params[1]
b = params[0]

sigma_a = np.sqrt(cov_mat[1,1])
sigma_b = np.sqrt(cov_mat[0,0])

print(f'Fit result: a:{a:.3f} ± {sigma_a:.3f} and b:{b:.3f} ± {sigma_b:.3f}')
```

```python
# Setze hier den Maximalen Wert der X-Werte ein:
x_max = 10.0
x_fine = np.linspace(0,x_max,num=100)
y_fit = np.polyval(params, x_fine)

plt.errorbar(array_x,array_y,yerr=array_y_std,marker='d', linestyle='none')
plt.plot(x_fine, y_fit)
plt.xlabel('x-achse')
plt.ylabel('y-achse')
plt.savefig('linear_fit.pdf')
```
